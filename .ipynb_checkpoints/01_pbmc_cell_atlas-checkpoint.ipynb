```python
import scanpy as sc

adata = sc.datasets.pbmc3k()

# QC
adata.var["mt"] = adata.var_names.str.startswith("MT-")
sc.pp.calculate_qc_metrics(adata, qc_vars=["mt"], inplace=True)
adata = adata[adata.obs.n_genes_by_counts > 200, :]
adata = adata[adata.obs.n_genes_by_counts < 2500, :]
adata = adata[adata.obs.pct_counts_mt < 5, :]

# Normalize + log
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# <-- TO JEST KLUCZ: zachowujemy pełne geny do markerów/plotów
adata.raw = adata

# HVG + PCA + UMAP + Leiden
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var.highly_variable]
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver="arpack")
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=10)
sc.tl.umap(adata)
sc.tl.leiden(adata, resolution=0.5)

sc.pl.umap(adata, color="leiden")
```

```python
markers = ["IL7R","CD3D","MS4A1","CD79A","NKG7","GNLY","LYZ","S100A8","S100A9","FCER1A","CST3","PPBP"]

sc.pl.dotplot(adata, markers, groupby="leiden", use_raw=True)
```

```python
celltype_map = {
    "0": "CD4 T cells",
    "1": "CD4 T cells",
    "2": "Unknown",
    "3": "B cells",
    "4": "NK cells",
    "5": "Unknown",
    "6": "Monocytes",
    "7": "Unknown",
    "8": "Dendritic cells",
    "9": "Platelets"
}

adata.obs["celltype"] = adata.obs["leiden"].map(celltype_map)
```

```python
sc.pl.umap(adata, color="celltype", legend_loc="on data")
```
